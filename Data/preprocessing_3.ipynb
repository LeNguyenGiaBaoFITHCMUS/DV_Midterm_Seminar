```python
import pandas as pd
```

```python
df_countries = pd.read_csv('production_countries.csv')
```

```python
df_countries.sample(10)
```

```python
# 1) Xác định tất cả các cột product_country_n...
country_cols = [col for col in df_countries.columns if col.startswith('product_country_n')]

# 2) Đếm số giá trị không NaN trên từng cột, sau đó cộng lại
total_non_nan = df_countries[country_cols].notnull().sum().sum()

print("Tổng số giá trị không NaN:", total_non_nan)
```

```python
df_countries[df_countries['id'] == 631534]
```

```python
# Lấy danh sách các cột quốc gia
country_cols = [col for col in df_countries.columns if col.startswith('product_country')]

# Sử dụng melt để chuyển đổi dataframe
df_long = df_countries.melt(id_vars=['id'], value_vars=country_cols,
                  var_name='variable', value_name='country')

# Loại bỏ các giá trị NaN
df_long = df_long.dropna(subset=['country'])

# Nếu không cần cột 'variable' nữa, bạn có thể loại bỏ nó
df_long = df_long[['id', 'country']]
```

```python
# Xóa các dòng trùng lặp
df_long = df_long.drop_duplicates()
```

```python
# Lưu trữ dataframe đã xử lý vào file CSV mới
df_long.to_csv('production_countries_long.csv', index=False)
```
