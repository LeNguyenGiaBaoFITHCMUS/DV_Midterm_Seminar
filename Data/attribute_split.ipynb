```python
import pandas as pd
```

```python
df = pd.read_csv('TMDB_movie_dataset_2000_2025.csv')
```

```python
df.columns
```

## Loại bỏ cột `overview` và cột `poster_path`
Cột `overview` là mô tả ngắn gọn về phim, có ý nghĩa về mặt ngôn ngữ tự nhiên, nhưng không thể trực quan hóa nếu không có độ đo cụ thể. Do đó, cột này sẽ được loại bỏ.  
Cột `poster_path` là đường dẫn đến ảnh bìa giới thiệu phim, không có ý nghĩa về mặt thống kê, do đó cũng được loại bỏ.

```python
df = df[['id', 'title', 'vote_average', 'vote_count', 'status', 'release_date', 'revenue', 'runtime', 'adult', 'budget', 'original_language',
       'original_title', 'popularity', 'genres', 'production_companies', 'production_countries', 'spoken_languages']]
df
```

## Tách các thể loại phim trong cột `genres`
Thuộc `genres` chứa nhiều giá trị, do đó chúng ta sẽ tách các giá trị của thuộc tính này ra. 

```python
genre_df = df[['id']]

for id in genre_df['id']:
    genres = df.loc[df['id'] == id, 'genres'].to_list()[0]
    if type(genres) != str:
        continue

    for genre in genres.split(', '):
        genre = genre.strip()
        if genre not in genre_df.keys():
            genre_df[genre] = False
        genre_df.loc[genre_df['id'] == id, genre] = True
```

```python
genre_df
```

```python
genre_df.to_csv('movie_genre_table.csv')
```

## Tách các nhà sản xuất phim trong cột `production_companies`
Thuộc `production_companies` cũng chứa nhiều giá trị, do đó chúng ta sẽ tách các giá trị của thuộc tính này ra.  

```python
product_companies_df = df[['id']]
for i in range(1, 16):
    product_companies_df[f'product_company_n{i}'] = pd.NA

for id in product_companies_df['id']:
    companies = df.loc[df['id'] == id, 'production_companies'].to_list()[0]
    if type(companies) != str:
        continue

    companies = companies.split(', ')
    for i in range(len(companies)):
        product_companies_df.loc[product_companies_df['id'] == id, f'product_company_n{i + 1}'] = companies[i].strip()
```

```python
product_companies_df
```

```python
product_companies_df.to_csv('production_companies.csv')
```

## Tách các nước sản xuất phim trong cột `production_countries`
Thuộc `production_countries` cũng chứa nhiều giá trị, do đó chúng ta sẽ tách các giá trị của thuộc tính này ra.  

```python
product_countries_df = df[['id']]
for i in range(1, 9):
    product_countries_df[f'product_country_n{i}'] = pd.NA

for id in product_companies_df['id']:
    countries = df.loc[df['id'] == id, 'production_countries'].to_list()[0]
    if type(countries) != str:
        continue

    countries = countries.split(', ')
    for i in range(len(countries)):
        product_countries_df.loc[product_countries_df['id'] == id, f'product_country_n{i + 1}'] = countries[i].strip()
```

```python
product_countries_df
```

```python
product_countries_df.to_csv('production_countries.csv')
```
