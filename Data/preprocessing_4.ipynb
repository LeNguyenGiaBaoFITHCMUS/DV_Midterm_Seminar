```python
import pandas as pd
```

```python
df_companies = pd.read_csv('production_companies.csv')
```

```python
df_companies.sample(10)
```

```python
df_companies[df_companies['id'] == 1303380]
```

```python
# 1) Xác định tất cả các cột product_country_n...
company_cols = [col for col in df_companies.columns if col.startswith('product_company_n')]

# 2) Đếm số giá trị không NaN trên từng cột, sau đó cộng lại
total_non_nan = df_companies[company_cols].notnull().sum().sum()

print("Tổng số giá trị không NaN:", total_non_nan)
```

```python
# Lấy danh sách các cột công ty
company_cols = [col for col in df_companies.columns if col.startswith('product_company')]

# Sử dụng melt để chuyển đổi dataframe
df_long = df_companies.melt(id_vars=['id'], value_vars=company_cols,
                  var_name='variable', value_name='country')

# Loại bỏ các giá trị NaN
df_long = df_long.dropna(subset=['country'])

# Nếu không cần cột 'variable' nữa, bạn có thể loại bỏ nó
df_long = df_long[['id', 'country']]
```

```python
# Xóa các dòng trùng lặp
df_long = df_long.drop_duplicates()
```

```python
# Lưu trữ dataframe đã xử lý vào file CSV mới
df_long.to_csv('production_companies_long.csv', index=False)
```
